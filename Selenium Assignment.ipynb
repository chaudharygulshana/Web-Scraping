**Q10: Write a python program to scrape the salary data for Data Scientist designation.
You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary.**
The above task will be, done as shown in the below steps:
1. First get the webpage https://www.ambitionbox.com/
2. Click on the salaries option as shown in the image.
3. After reaching to the following webpage, In place of “Search Job Profile” enters “Data Scientist” and
then click on “Data Scientist”.<br>
You have to scrape the data ticked in the above image.<br>
4. Scrape the data for the first 10 companies. Scrape the company name, total salary record, average
salary, minimum salary, maximum salary, experience required.
5. Store the data in a dataframe.


```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get(" https://www.ambitionbox.com/")
```

```python
#Click on salaries option
nxt_btn=driver.find_element(By.XPATH,"/html/body/div[1]/nav[2]/div/ul/li[3]/a")
nxt_btn.click()
```

```python
#Click on browse salaries option
nxt_btn=driver.find_element(By.XPATH,"/html/body/div[1]/nav[2]/div/ul/li[3]/div/ul/li[1]/div")
nxt_btn.click()
```

```python
# Search for Data Scientist
srch=driver.find_element(By.XPATH,"/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
srch.send_keys("Data Scientist")
```

```python

#Click on DATA SCIENTIST option
nxt_btn=driver.find_element(By.XPATH,"/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p")
nxt_btn.click()
```

```python
company_info=[i.text for i in driver.find_elements(By.XPATH,"//div[@class='company-info']/a")][:10]
# Company
company_name=[]
for i in range(len(company_info)):
    x=company_info[i].split('\n')[0]
    company_name.append(x)
# Exp required
exp=[i.text for i in driver.find_elements(By.XPATH,"//div[@class='sbold-list-header']")][:10]  
exp[1].split('(')
experience=[]
for i in range(len(exp)):
    x=exp[i].split('(')[0]
    experience.append(x)


# Based On salaries
salary_based=[i.text for i in driver.find_elements(By.XPATH,"//div[@class='sbold-list-header']/span")][:10] 

# find minimum and maximum salary
min_salary=[]
max_salary=[]
salary=[i.text for i in driver.find_elements(By.XPATH,"//div[@class='salary-values']")][:10] 
for i in range(len(salary)):
    x=salary[i].split('\n')[0]
    min_salary.append(x)
    y=salary[i].split('\n')[1]
    max_salary.append(y)
# Average Salary
avg_salary=[i.text for i in driver.find_elements(By.XPATH,"//p[@class='averageCtc']")][:10]

df=pd.DataFrame({'Company':company_name,'Experience':experience,'Based on Salaries':salary_based,'Minimum Salary':min_salary,'Maximum Salary':max_salary,"Average Salary":avg_salary})
df
```
